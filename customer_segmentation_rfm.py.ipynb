```python
import pandas as pd
df=pd.read_excel("QVI_transaction_data.xlsx")
df

# df1=df.isnull().sum()       
# df1.duplicated()
# df1
```

```python
df.info()
```

```python
# print(type(df1))
# print(df1["PROD_NAME"].dtype)
df["brand"]=df["PROD_NAME"].str.split().str[0]
df
```

```python
df["packsize"]=df["PROD_NAME"].str.extract(r"(\d+g)")
df
```

```python
# df.drop("packsize",axis=1,inplace=True )
# df
```

```python
df["DATE"]=pd.to_datetime(df["DATE"])
df
```

```python
customermetrics=df.groupby("LYLTY_CARD_NBR").agg(Total_spend=("TOT_SALES","sum"),
purchase_frequency=("TXN_ID","nunique"),avg_spend_per_transaction=("TOT_SALES","mean")).reset_index()
customermetrics
```

```python
for name, group in df.groupby("LYLTY_CARD_NBR"):
    if group["packsize"].mode().empty or group["brand"].mode().empty:
        print(f"Group {name} has an issue: {group}")
```

```python
df.empty
```

```python
prefer_pack_brand=df.groupby("LYLTY_CARD_NBR").agg(prefer_pack=("packsize",lambda x:x.mode().iloc[0] 
if not x.mode().empty else None),prefer_brand=("brand",lambda x:x.mode().iloc[0] if not x.mode().empty else None)).reset_index()
prefer_pack_brand
```

```python
customermetric=customermetrics.merge(prefer_pack_brand,on="LYLTY_CARD_NBR")
customermetric    

# customermetrics.info(
```

```python
df["DATE"]=pd.to_datetime(df["DATE"])
df
```

```python
df["days_since_purchase"]=(df["DATE"].max()-df["DATE"]).dt.days
df

rfm=df.groupby("LYLTY_CARD_NBR").agg(Recency=("days_since_purchase","min"),frequency=("TXN_ID","nunique"),monetery=("TOT_SALES","sum"))
rfm
```

```python
from sklearn.preprocessing import StandardScaler 
scalar=StandardScaler ()
rfm_scalar=scalar.fit_transform(rfm[["Recency","frequency","monetery"]])
rfm_scalar
```

```python
import numpy as np
from sklearn.cluster import KMeans
kmean=KMeans(n_clusters=3,random_state=2)
rfm["segment"]=kmean.fit_predict (rfm_scalar)
rfm
unique_clusters, counts = np.unique(rfm["segment"], return_counts=True)

print(f"Number of clusters: {len(unique_clusters)}")
print(f"Cluster assignments: {unique_clusters}")
print(f"Points per cluster: {dict(zip(unique_clusters, counts))}")
```

```python
top_brand=customermetric .groupby("prefer_brand").agg(total_spend=("Total_spend","sum")).reset_index()
# print (top_brand.columns)
top_brand=top_brand.sort_values(by="total_spend",ascending=False)
top_brand
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(17,8))    
sns.barplot(x="prefer_brand",y="total_spend",data=top_brand)
plt.xticks(rotation=45, fontsize=12, ha='center') 
plt.title("Total spend")
plt.xlabel("prefer_brand")
plt.ylabel("total_spend")
plt.tight_layout()
plt.show()
plt.savefig("total _spend.png")
```

```python
plt.figure(figsize=(10,6))    
sns.scatterplot(data=rfm,x="frequency",y="monetery",hue='segment', palette='Set2')
plt.title("customer segment")
plt.xlabel("frequency")
plt.ylabel("monetery")
plt.tight_layout()

plt.show()
```
